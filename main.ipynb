```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load dataset
df = pd.read_csv("perfume.csv")
df.head()
```

```python
df.info()
```

```python
# drop duplicate data
duplicate = df[df.duplicated()]
print(duplicate.shape)

df = df.drop_duplicates()
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python
# displaying data with barplot
few = df.head(20)
plt.bar(few["brand"], few["followers"], color="blue", width=0.5)
plt.xlabel("Brand Parfum Indonesia")
plt.ylabel("Pengikut Media Sosial")
plt.title("Data Jumlah Pengikut Media Sosial Brand Parfum Indonesia")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# displaying data with pieplot
few = df.head(20)
plt.pie(few["followers"], labels=few["brand"])
```

```python
# displaying data with scatterplot
few = df.head(20)
plt.scatter(few["followers"], few["price_per_ml"])
plt.xlabel("Followers")
plt.ylabel("Price per mL")
plt.title("Scatter Plot of Followers vs. Price per mL")
plt.show()
```

```python
# display followers outliers
outl = plt.figure(figsize=(12,4))
outl.add_subplot(1,2,1)
df['followers'].plot(kind='kde')
outl.add_subplot(1,2,2)
plt.boxplot(df['followers'])
plt.show()
```
